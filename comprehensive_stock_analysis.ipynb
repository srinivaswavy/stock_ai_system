## üìä Price Analysis & Statistics

Historical price data and statistical analysis including returns, volatility, and technical indicators.

```python
# Price Analysis and Statistics
print(f"üìä PRICE ANALYSIS & STATISTICS FOR {STOCK_SYMBOL}")
print("=" * 60)

# Get price statistics for different periods
periods = ["1mo", "3mo", "6mo", "1y"]
price_stats = {}

for period in periods:
    stats = analyzer.get_price_statistics(period=period)
    if stats:
        price_stats[period] = stats
        print(f"\nüìà {period.upper()} Statistics:")
        print(f"   Current Price: ${stats.get('current_price', 'N/A')}")
        print(f"   Period High: ${stats.get('period_high', 'N/A')}")
        print(f"   Period Low: ${stats.get('period_low', 'N/A')}")
        print(f"   Price Change: {stats.get('change_pct', 'N/A'):.2f}%" if stats.get('change_pct') else "   Price Change: N/A")
        print(f"   Volatility: {stats.get('volatility', 'N/A'):.2f}%" if stats.get('volatility') else "   Volatility: N/A")

# Get recent historical data
print(f"\nüìã RECENT TRADING DATA:")
print("-" * 40)
recent_data = analyzer.get_historical_data(period="5d")
if recent_data is not None and not recent_data.empty:
    # Display last 5 days
    display(recent_data.tail())
    
    # Calculate daily statistics
    latest_close = recent_data['Close'].iloc[-1]
    daily_change = recent_data['Close'].pct_change().iloc[-1] * 100
    avg_volume = recent_data['Volume'].mean()
    
    print(f"\nüìä Latest Trading Metrics:")
    print(f"   Latest Close: ${latest_close:.2f}")
    print(f"   Daily Change: {daily_change:+.2f}%")
    print(f"   Average Volume: {avg_volume:,.0f}")
else:
    print("‚ùå Could not fetch recent trading data")

print("=" * 60)
```

## üì∞ Enhanced News Coverage

Comprehensive news analysis with multiple retrieval methods, full content extraction, and detailed statistics.

```python
# Ultra-Comprehensive News Analysis
print(f"üì∞ ULTRA-COMPREHENSIVE NEWS ANALYSIS FOR {STOCK_SYMBOL}")
print("=" * 70)

# Get ultra-comprehensive news data
ultra_news = analyzer.get_ultra_comprehensive_news(limit=40)

if ultra_news.get('status') == 'success':
    coverage = ultra_news['coverage_summary']
    quality = ultra_news['quality_metrics']
    content_analysis = ultra_news['content_analysis']
    
    # Display comprehensive statistics
    print(f"üìä NEWS COVERAGE STATISTICS:")
    print(f"   üì∞ Total articles available: {ultra_news['total_available_articles']}")
    print(f"   ‚úÖ Articles processed: {ultra_news['total_processed_articles']}")
    print(f"   üìù Articles with content: {coverage['articles_with_content']}")
    print(f"   üíé Substantial articles: {coverage['substantial_articles']}")
    print(f"   üìä Total content: {coverage['total_characters']:,} characters")
    print(f"   üî§ Total words: {coverage['total_words']:,}")
    print(f"   üè¢ Unique publishers: {coverage['unique_publishers']}")
    print(f"   üìà Coverage quality: {quality['coverage_ratio']:.1%}")
    print(f"   üìè Avg article length: {quality['avg_content_length']:.0f} characters")
    
    # Display top 10 news articles
    articles_with_content = [a for a in ultra_news['articles'] if not a.get('error') and a.get('char_count', 0) > 0]
    articles_with_content.sort(key=lambda x: x.get('char_count', 0), reverse=True)
    
    print(f"\nüèÜ TOP 10 MOST COMPREHENSIVE NEWS ARTICLES:")
    print("-" * 70)
    
    for i, article in enumerate(articles_with_content[:10], 1):
        print(f"\n{i:2d}. {article['title']}")
        print(f"    üìä {article['char_count']:,} chars | {article['word_count']:,} words")
        print(f"    üè¢ {article['publisher']}")
        print(f"    üìÖ {article['published_date']}")
        print(f"    üîó {article['link']}")
        
        # Show content preview
        if article.get('full_content'):
            preview = article['full_content'][:200].replace('\n', ' ').strip()
            if len(article['full_content']) > 200:
                preview += "..."
            print(f"    üìù {preview}")
        
        # Show sentiment indicators if significant
        sentiment = article.get('sentiment_indicators', {})
        if sum(sentiment.values()) > 0:
            sentiment_str = " | ".join([f"{k}: {v}" for k, v in sentiment.items() if v > 0])
            print(f"    üí≠ Sentiment: {sentiment_str}")
    
    # Publisher breakdown
    if content_analysis['unique_publishers']:
        print(f"\nüì∞ NEWS PUBLISHERS ({len(content_analysis['unique_publishers'])} total):")
        print("-" * 50)
        for i, publisher in enumerate(content_analysis['unique_publishers'][:15], 1):
            print(f"{i:2d}. {publisher}")
        if len(content_analysis['unique_publishers']) > 15:
            print(f"    ... and {len(content_analysis['unique_publishers']) - 15} more")
            
else:
    print(f"‚ùå Could not fetch news data: {ultra_news.get('error', 'Unknown error')}")

print("\n" + "=" * 70)
```

```python
# News Methods Comparison
print(f"üîç NEWS METHODS COMPARISON FOR {STOCK_SYMBOL}")
print("=" * 60)

# Compare all available news methods
comparison = analyzer.compare_news_methods([STOCK_SYMBOL])

if STOCK_SYMBOL in comparison:
    methods = comparison[STOCK_SYMBOL]
    
    print(f"üìä METHOD COMPARISON RESULTS:")
    print("-" * 50)
    
    # Display each method's results
    for method_name, method_data in methods.items():
        method_display = method_name.replace('_', ' ').title()
        articles = method_data.get('articles', 0)
        total_chars = method_data.get('total_characters', 0)
        avg_chars = method_data.get('avg_chars_per_article', 0)
        
        print(f"\nüîπ {method_display}:")
        print(f"   Articles: {articles}")
        print(f"   Total Content: {total_chars:,} characters")
        print(f"   Average per Article: {avg_chars:.0f} characters")
        
        if method_name == 'ultra_comprehensive_method':
            publishers = method_data.get('unique_publishers', 0)
            quality = method_data.get('coverage_quality', 0)
            print(f"   Unique Publishers: {publishers}")
            print(f"   Coverage Quality: {quality:.1%}")
    
    # Find the best method
    best_total_chars = 0
    best_method = None
    for method_name, method_data in methods.items():
        if method_data.get('total_characters', 0) > best_total_chars:
            best_total_chars = method_data.get('total_characters', 0)
            best_method = method_name
    
    if best_method:
        print(f"\nüèÜ BEST METHOD: {best_method.replace('_', ' ').title()}")
        print(f"   Provides {best_total_chars:,} characters of news content")
else:
    print(f"‚ùå Could not compare news methods for {STOCK_SYMBOL}")

print("\n" + "=" * 60)
```

## üìã Financial Metrics & Recommendations

Analyst recommendations, calendar events, dividend information, and other financial metrics.

```python
# Financial Metrics and Recommendations
print(f"üìã FINANCIAL METRICS & RECOMMENDATIONS FOR {STOCK_SYMBOL}")
print("=" * 65)

# Get analyst recommendations
print("üí° ANALYST RECOMMENDATIONS:")
print("-" * 40)
recommendations = analyzer.get_stock_recommendations()
if recommendations is not None and not recommendations.empty:
    display(recommendations.head(10))
else:
    print("‚ùå No analyst recommendations available")

# Get calendar events
print(f"\nüìÖ UPCOMING CALENDAR EVENTS:")
print("-" * 40)
calendar = analyzer.get_calendar_events()
if calendar is not None:
    if isinstance(calendar, dict):
        for key, value in calendar.items():
            print(f"   {key}: {value}")
    elif hasattr(calendar, 'empty') and not calendar.empty:
        display(calendar)
    else:
        print("   No upcoming calendar events found")
else:
    print("   No calendar events available")

# Get dividend information
print(f"\nüí∞ DIVIDEND INFORMATION:")
print("-" * 40)
dividends = analyzer.get_dividend_data()
if dividends is not None and not dividends.empty:
    print(f"   Recent dividends (last 10):")
    recent_dividends = dividends.tail(10)
    for date, amount in recent_dividends.items():
        print(f"   {date.strftime('%Y-%m-%d')}: ${amount:.2f}")
    
    # Calculate dividend statistics
    annual_dividend = dividends.tail(4).sum() if len(dividends) >= 4 else dividends.sum()
    print(f"\n   üíé Estimated Annual Dividend: ${annual_dividend:.2f}")
    
    if stock_info and stock_info.get('Current Price'):
        try:
            current_price = float(str(stock_info['Current Price']).replace('$', ''))
            dividend_yield = (annual_dividend / current_price) * 100
            print(f"   üìä Dividend Yield: {dividend_yield:.2f}%")
        except:
            print("   üìä Dividend Yield: Could not calculate")
else:
    print("   No dividend data available")

print("\n" + "=" * 65)
```

```python
# Comprehensive Analysis Summary
print(f"üîç COMPREHENSIVE ANALYSIS SUMMARY FOR {STOCK_SYMBOL}")
print("=" * 65)

# Get comprehensive analysis
comprehensive_analysis = analyzer.get_comprehensive_analysis(period="6mo", include_news=True)

if comprehensive_analysis:
    # Stock Information Summary
    stock_summary = comprehensive_analysis.get('stock_info', {})
    print(f"üìä STOCK SUMMARY:")
    print(f"   Company: {stock_summary.get('Company Name', 'N/A')}")
    print(f"   Price: ${stock_summary.get('Current Price', 'N/A')}")
    print(f"   Market Cap: {stock_summary.get('Market Cap', 'N/A')}")
    print(f"   Sector: {stock_summary.get('Sector', 'N/A')}")
    
    # Technical Analysis
    technical = comprehensive_analysis.get('technical_analysis', {})
    print(f"\nüìà TECHNICAL ANALYSIS:")
    print(f"   Trend Signal: {technical.get('trend_signal', 'N/A')}")
    print(f"   RSI: {technical.get('rsi', 'N/A')}")
    print(f"   Support Level: ${technical.get('support_level', 'N/A')}")
    print(f"   Resistance Level: ${technical.get('resistance_level', 'N/A')}")
    
    # Price Statistics
    price_analysis = comprehensive_analysis.get('price_analysis', {})
    print(f"\nüí∞ PRICE ANALYSIS (6 months):")
    print(f"   Current Price: ${price_analysis.get('current_price', 'N/A')}")
    print(f"   Period Return: {price_analysis.get('change_pct', 'N/A'):.2f}%" if price_analysis.get('change_pct') else "   Period Return: N/A")
    print(f"   Volatility: {price_analysis.get('volatility', 'N/A'):.2f}%" if price_analysis.get('volatility') else "   Volatility: N/A")
    print(f"   Period High: ${price_analysis.get('period_high', 'N/A')}")
    print(f"   Period Low: ${price_analysis.get('period_low', 'N/A')}")
    
    # News Summary
    news_summary = comprehensive_analysis.get('news_summary', {})
    print(f"\nüì∞ NEWS SUMMARY:")
    print(f"   Total Articles: {news_summary.get('total_articles', 'N/A')}")
    print(f"   Recent Headlines: {news_summary.get('recent_headlines', 'N/A')}")
    
    # Investment Recommendation (based on technical analysis)
    trend_signal = technical.get('trend_signal', '').lower()
    rsi = technical.get('rsi', 0)
    
    print(f"\nüéØ ANALYSIS CONCLUSION:")
    print(f"   Trend: {trend_signal.title() if trend_signal else 'Neutral'}")
    
    if rsi:
        if rsi > 70:
            rsi_signal = "Overbought (Consider Selling)"
        elif rsi < 30:
            rsi_signal = "Oversold (Consider Buying)"
        else:
            rsi_signal = "Neutral"
        print(f"   RSI Signal: {rsi_signal}")
    
    if trend_signal == 'bullish' and rsi < 70:
        conclusion = "üü¢ POSITIVE - Bullish trend with reasonable RSI"
    elif trend_signal == 'bearish' and rsi > 30:
        conclusion = "üî¥ NEGATIVE - Bearish trend"
    else:
        conclusion = "üü° NEUTRAL - Mixed signals"
    
    print(f"   Overall: {conclusion}")
    
else:
    print("‚ùå Could not generate comprehensive analysis")

print(f"\nüìä ANALYSIS TIMESTAMP: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print("=" * 65)
```

## üíæ Data Export & Summary

Export all analysis data to JSON files and display final summary.

```python
# Data Export and Final Summary
print(f"üíæ DATA EXPORT & FINAL SUMMARY FOR {STOCK_SYMBOL}")
print("=" * 60)

# Generate timestamp for file naming
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

try:
    # Export comprehensive analysis data
    print("üìÅ Exporting analysis data...")
    
    # 1. Export ultra-comprehensive news data
    if ultra_news.get('status') == 'success':
        news_filename = f"{STOCK_SYMBOL}_comprehensive_news_{timestamp}.json"
        with open(news_filename, 'w', encoding='utf-8') as f:
            json.dump(ultra_news, f, indent=2, ensure_ascii=False)
        print(f"   ‚úÖ News data exported: {news_filename}")
    
    # 2. Export complete stock analysis
    complete_analysis = {
        'symbol': STOCK_SYMBOL,
        'analysis_timestamp': datetime.now().isoformat(),
        'stock_info': stock_info,
        'price_statistics': price_stats,
        'comprehensive_analysis': comprehensive_analysis,
        'news_methods_comparison': comparison.get(STOCK_SYMBOL, {}),
        'data_sources': [
            'Yahoo Finance API',
            'Enhanced StockAnalyzer',
            'ComprehensiveNewsAnalyzer'
        ]
    }
    
    analysis_filename = f"{STOCK_SYMBOL}_complete_analysis_{timestamp}.json"
    with open(analysis_filename, 'w', encoding='utf-8') as f:
        json.dump(complete_analysis, f, indent=2, ensure_ascii=False, default=str)
    print(f"   ‚úÖ Complete analysis exported: {analysis_filename}")
    
    # 3. Export using the comprehensive news analyzer
    comprehensive_news_file = news_analyzer.export_news_data(yahoo_limit=35)
    print(f"   ‚úÖ Comprehensive news exported: {comprehensive_news_file}")
    
except Exception as e:
    print(f"   ‚ùå Export error: {e}")

# Final Summary
print(f"\nüéâ COMPREHENSIVE STOCK ANALYSIS COMPLETE!")
print("=" * 60)
print(f"üìä Stock Analyzed: {STOCK_SYMBOL}")
print(f"‚è∞ Analysis Time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

if ultra_news.get('status') == 'success':
    coverage = ultra_news['coverage_summary']
    print(f"üì∞ News Articles Analyzed: {coverage['articles_with_content']}")
    print(f"üìù Total News Content: {coverage['total_characters']:,} characters")
    print(f"üè¢ News Publishers: {coverage['unique_publishers']}")

if stock_info:
    print(f"üí∞ Current Price: ${stock_info.get('Current Price', 'N/A')}")
    print(f"üìä Market Cap: {stock_info.get('Market Cap', 'N/A')}")

print(f"\nüìÅ Exported Files:")
try:
    print(f"   ‚Ä¢ {news_filename}")
    print(f"   ‚Ä¢ {analysis_filename}")  
    print(f"   ‚Ä¢ {comprehensive_news_file}")
except:
    print(f"   ‚Ä¢ Check directory for exported JSON files")

print(f"\n‚ú® ANALYSIS FEATURES USED:")
print(f"   ‚úÖ Ultra-comprehensive news analysis")
print(f"   ‚úÖ Multiple news retrieval methods")
print(f"   ‚úÖ Technical analysis and indicators")
print(f"   ‚úÖ Interactive price charts")
print(f"   ‚úÖ Financial metrics and recommendations")
print(f"   ‚úÖ Complete data export")

print(f"\nüîÑ To analyze a different stock:")
print(f"   1. Change the STOCK_SYMBOL variable at the top")
print(f"   2. Run all cells again")

print("\n" + "=" * 60)
print("üéØ READY FOR NEXT ANALYSIS!")
```
